```python
df1 = pd.DataFrame({
    "ID":[1,2,3,4],
    "Name":["A","B","C","D"],
    "Class":[5,6,7,8]
})
df1
```

```python
df2 = pd.DataFrame({
    "ID":[5,6,7,8],
    "Name":["E","F","G","H"],
    "Class":[9,10,11,12]
})
df2
```

```python
df1 = pd.DataFrame({
    "ID":[1,2,3,4],
    "Name":["A","B","C","D"],
    "Class":[5,6,7,8]
})
df1
```

```python
df2 = pd.DataFrame({
    "ID":[3,4],
    "Name":["C","D"],
    "Class":[7,8]
})
df2
```

```python
pd.concat([df1,df2],axis=1,ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
pd.concat([df1,df2],axis=1,join_axes=[df1.index])
```

```python
pd.concat([df1,df2],axis=1,join_axes=[df2.index])
```
