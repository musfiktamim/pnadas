```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "ID":[1,2,3,5],
    "Class":[9,10,11,12]
})
df1
```

```python
df2 = pd.DataFrame({
    "ID":[1,2,3,4],
    "Name":["A","B","C","D"]
})
df2
```

```python
pd.merge(df2,df1,on="ID")
```

```python
pd.merge(df2,df1,on="ID",how="right")
```

```python
pd.merge(df2,df1,on="ID",how="left")
```

```python
pd.merge(df2,df1,on="ID",how="outer",)
```

```python
pd.merge(df2,df1,on="ID",how="outer",indicator=True)
```

```python
df2 = pd.DataFrame({
    "ID":[5,6,7,8],
    "Name":["A","B","C","D"]
})
df2
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
df1 = pd.DataFrame({
    "ID":[1,2,3,5],
    "Class":[9,10,11,12]
})
df1
```

```python
df2 = pd.DataFrame({
    "ID":[1,2,3,4],
    "Class":[9,10,11,12]
})
df2
```

```python
pd.merge(df2,df1,on="ID")
```

```python
pd.merge(df2,df1,on="ID",suffixes=("Higher","Middle"))
```
